```python

import requests
import openpyxl

# Chave de API
API_KEY = ''

# URL base da API de Distância da Matriz do DistanceMatrix.ai
base_url = "https://api.distancematrix.ai/maps/api/distancematrix/json"

# Ponto de origem
origem = "Pr. Almirante Gago Coutinho, 29 - Ponta da Praia, Santos - SP, 11030-200"


# Dicionário para armazenar as distâncias
distancias = {}

# Calcular distância para cada cidade
for cidade in cidades:
    try:
        # Parâmetros da solicitação
        params = {
            'origins': origem,
            'destinations': cidade,
            'key': API_KEY
        }

        # Faz a solicitação à API
        response = requests.get(base_url, params=params)
        data = response.json()

        # Extrai a distância do resultado
        distancia_texto = data['rows'][0]['elements'][0]['distance']['text']
        distancia_valor = data['rows'][0]['elements'][0]['distance']['value']

        # Armazena a distância no dicionário
        distancias[cidade] = {
            'distancia_texto': distancia_texto,
            'distancia_valor': distancia_valor
        }
        
    except Exception as e:
        print(f"Erro ao calcular distância para {cidade}: {e}")

# Salvar os resultados em um arquivo Excel
wb = openpyxl.Workbook()
sheet = wb.active

# Adiciona cabeçalhos
sheet['A1'] = 'Cidade'
sheet['B1'] = 'Distância'

# Adiciona os dados das distâncias
row = 2
for cidade, distancia_info in distancias.items():
    sheet[f'A{row}'] = cidade
    sheet[f'B{row}'] = distancia_info['distancia_texto']
    row += 1

# Salva o arquivo Excel
wb.save('distancias.xlsx')
```
